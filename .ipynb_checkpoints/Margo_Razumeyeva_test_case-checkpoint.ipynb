```python
def stadartise_col_name(df):
    """Change columns name to snake_case"""
    new_cols_names = [re.sub(r"(?<=\w)([A-Z])", r"_\1", col_name).lower() for col_name in list(df.columns)]

    df.columns = new_cols_names
    return df

def print_dates(df, date_column='full_date', title='communications per month'):
    """Print the number of communication per months"""
    print("Dataframe cover dates")
    print(f'from {df[date_column].min()}\t to {df[date_column].max()}')

    df.index = pd.to_datetime(df[date_column], format='%Y-%m-%d')  #format='%m/%d/%y %I:%M%p')
    
    df['counter'] = 1
    df_gr =  df.groupby(pd.Grouper(freq='M'))['counter'].sum()
    
    fig, ax = plt.subplots()
    df_gr.plot(kind='bar', title=title)
    ax.set_xticklabels(list(pd.DataFrame(df_gr).reset_index()[date_column].dt.date))
    
    df.drop(['counter'], axis=1, inplace=True)
    return 
```



### hypothesis: it is possible to distinguish 4 groups of users who will have a different number of reservations

if it's true then the number of connections definitely affects the reservations.

```python
comm_col_name = 'connections'

# set persentille borders based on plot
border_1_pr = 0.4
border_2_pr = 0.75
border_3_pr = 0.9

# get values for each border
border_1_value = np.quantile(df_mer[comm_col_name], border_1_pr)
border_2_value = np.quantile(df_mer[comm_col_name], border_2_pr)
border_3_value = np.quantile(df_mer[comm_col_name], border_3_pr)


conditions = [
    (df_mer[comm_col_name]  <= border_1_value),
    (df_mer[comm_col_name] > border_1_value) & (df_mer[comm_col_name] <= border_2_value),
    (df_mer[comm_col_name] > border_2_value) & (df_mer[comm_col_name] < border_3_value),
    (df_mer[comm_col_name] >= border_3_value)
    ]

# create a list of the values we want to assign for each condition
values = [1, 2, 3, 4]

# create a new column 
df_mer['group'] = np.select(conditions, values)

df_mer['group'].value_counts()
```

```python
for group_number in list(df_mer['group'].unique()):
    print(f"group number {group_number}")
    
    df_tmp = df_mer[df_mer['group'] == group_number]
    print(f"Total users {len(df_tmp)}")
    print("Mean reservations %.2f" % df_tmp['reservation'].mean())
    print("Mean connections %.2f" % df_tmp['connections'].mean())
    print(f"Total reservation {df_tmp['reservation'].sum()}, total connections {df_tmp['connections'].sum()}")
    print("")
```

### Confidence intervals

I calculate mean within the bootstrapping and build Confidence intervals for each group

```python
def get_bootstrap_samples(data, n_samples, n_subsets=500):
    """Bootstrapping functions - subsets with returns"""
    indices = np.random.randint(0, len(data), (n_samples, n_subsets))
    samples = data[indices]
    return samples


def stat_intervals(stat, alpha):
    boundaries = np.percentile(stat, [100 * alpha / 2., 100 * (1 - alpha / 2.)])
    return boundaries


def plot_confidence_interval(x, mean, top, bottom, color='#2187bb', horizontal_line_width=0.25):
    left = x - horizontal_line_width / 2
    right = x + horizontal_line_width / 2

    plt.plot([x, x], [top, bottom], color=color)
    plt.plot([left, right], [top, top], color=color)
    plt.plot([left, right], [bottom, bottom], color=color)
    plt.plot(x, mean, 'o', color='#f44336')

    return 
```

```python
df_result = df_mer.groupby('group').mean()
df_result['left_border_res'] = 0
df_result['right_border_res'] = 0

np.random.seed(42)
for i in list(df_mer['group'].unique()):
    gr_mean = get_bootstrap_samples(df_mer[df_mer['group'] == i]['reservation'].values, 200).mean(axis=0)

    
    lb, rb = stat_intervals(gr_mean, 0.05)
    df_result.loc[i, 'left_border_res'] = lb
    df_result.loc[i, 'right_border_res'] = rb
```


Confidence intervals do not overlaps, so we can conclude that number of connections affects number of reservations.

Let's plot it.

```python
for i in range(1, 5):
    plot_confidence_interval(i, df_result.loc[i, 'reservation'] , 
                             df_result.loc[i, 'left_border_res'], 
                             df_result.loc[i, 'right_border_res'],
                             color='#2187bb', horizontal_line_width=0.25)
plt.title('confidence intervals')
plt.savefig('../img/conf.png')
```
